```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
wage_dict =     {"gamma_0": 10,
    "gamma_1": 4,
    "gamma_2": -0.08,}
```

```python
exp_levels = np.arange(0, 40)
```

```python
wage = wage_dict["gamma_0"] + wage_dict["gamma_1"] * exp_levels + (exp_levels ** 2) * wage_dict["gamma_2"]
```

```python
plt.plot(exp_levels, wage)
```
